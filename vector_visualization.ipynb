# calculating Determinant and Visualizing it

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the 2x2 matrix A
A = np.array([[0.5, -0.5],
              [0.5,  0.5]])

# Extract column vectors
v1 = A[:, 0]
v2 = A[:, 1]

# Calculate the determinant
det_A = np.linalg.det(A)

# Setup plot
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

# Plot the vectors
origin = np.array([0, 0])
ax.quiver(*origin, *v1, angles='xy', scale_units='xy', scale=1, color='r', label='v1')
ax.quiver(*origin, *v2, angles='xy', scale_units='xy', scale=1, color='b', label='v2')

# Draw the parallelogram
parallelogram = np.column_stack((origin, v1, v1 + v2, v2, origin))
ax.plot(parallelogram[0], parallelogram[1], 'g-', label='Parallelogram')

# Display determinant info
ax.set_title(f"Determinant = {det_A:.2f} (Signed Area)")

# Legend and show
ax.legend()
plt.show()
```

# V1 =[1,3] V2 =[2,4] = -2 visualizing

```python
import numpy as np
import matplotlib.pyplot as plt

# Define 2x2 matrix A
A = np.array([[1, 2],
              [3, 4]])

# Extract column vectors
v1 = A[:, 0]  # [1, 3]
v2 = A[:, 1]  # [2, 4]

# Compute determinant
det_A = np.linalg.det(A)

# Define origin and points for parallelogram
origin = np.array([0, 0])
p1 = v1
p2 = v2
p3 = v1 + v2

# Coordinates for the parallelogram (closed shape)
parallelogram = np.column_stack((origin, p1, p3, p2, origin))

# Create figure and 2 side-by-side subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
for ax in (ax1, ax2):
    ax.set_aspect('equal')
    ax.set_xlim(-1, 4)
    ax.set_ylim(-1, 6)
    ax.grid(True, linestyle='--', alpha=0.5)
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)

# Plot on ax1: vectors and parallelogram outline
ax1.quiver(*origin, *v1, angles='xy', scale_units='xy', scale=1, color='r', label='v1 = [1, 3]')
ax1.quiver(*origin, *v2, angles='xy', scale_units='xy', scale=1, color='b', label='v2 = [2, 4]')
ax1.plot(parallelogram[0], parallelogram[1], 'g-', label='Parallelogram')
ax1.set_title("Vectors and Parallelogram")
ax1.legend()

# Plot on ax2: parallelogram filled with color to show area
ax2.fill(parallelogram[0], parallelogram[1], color='lightblue', edgecolor='black')
ax2.text(1.5, 2.5, f"Area = {det_A:.2f}", fontsize=14, color='black', ha='center')
ax2.set_title("Determinant as Signed Area")

# Show plot
plt.tight_layout()
plt.show()
```

# 3D determinate 

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Define the 3x3 matrix
A = np.array([[1.0, 0.5, 1.0],
              [0.0, 1.0, 0.0],
              [1.0, 1.5, 1.0]])

# Extract column vectors
v1 = A[:, 0]
v2 = A[:, 1]
v3 = A[:, 2]

# Calculate determinant (signed volume)
det = np.linalg.det(A)

# Origin
O = np.array([0, 0, 0])

# Build 3D parallelepiped from origin using v1, v2, v3
points = [
    O,
    v1,
    v2,
    v3,
    v1 + v2,
    v1 + v3,
    v2 + v3,
    v1 + v2 + v3
]

# Define the 6 faces of the parallelepiped (each is a quadrilateral)
faces = [
    [points[0], points[1], points[4], points[2]],
    [points[0], points[2], points[6], points[3]],
    [points[0], points[3], points[5], points[1]],
    [points[7], points[6], points[2], points[4]],
    [points[7], points[5], points[1], points[4]],
    [points[7], points[5], points[3], points[6]],
]

# Plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the vectors
ax.quiver(*O, *v1, color='r', label='v1', arrow_length_ratio=0.1)
ax.quiver(*O, *v2, color='g', label='v2', arrow_length_ratio=0.1)
ax.quiver(*O, *v3, color='b', label='v3', arrow_length_ratio=0.1)

# Draw parallelepiped faces
ax.add_collection3d(Poly3DCollection(faces, facecolors='lightblue', edgecolors='k', linewidths=1, alpha=0.4))

# Axes limits
ax.set_xlim([0, 2])
ax.set_ylim([0, 2])
ax.set_zlim([0, 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Title
ax.set_title(f"3D Parallelepiped from Matrix Columns\nDeterminant (Signed Volume) = {det:.2f}")

# Legend and show
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Visualizing the matrix of above 
import numpy as np
import matplotlib.pyplot as plt

# Define matrices
A = np.array([[2, -3], [7, 5]])
B = np.array([[10, -8], [12, -2]])

# Matrix multiplication
C = np.dot(A, B)

C = C/4

# Create a figure
plt.figure(figsize=(8, 8))
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)

# Plot vectors from A
plt.quiver(0, 0, A[0,0], A[1,0], angles='xy', scale_units='xy', scale=1, color='blue', label='A column 1')
plt.quiver(0, 0, A[0,1], A[1,1], angles='xy', scale_units='xy', scale=1, color='blue', alpha=0.5, label='A column 2')

# Plot vectors from B
plt.quiver(0, 0, B[0,0], B[1,0], angles='xy', scale_units='xy', scale=1, color='green', label='B column 1')
plt.quiver(0, 0, B[0,1], B[1,1], angles='xy', scale_units='xy', scale=1, color='green', alpha=0.5, label='B column 2')

# Plot vectors from C
plt.quiver(0, 0, C[0,0], C[1,0], angles='xy', scale_units='xy', scale=1, color='red', label='C column 1')
plt.quiver(0, 0, C[0,1], C[1,1], angles='xy', scale_units='xy', scale=1, color='red', alpha=0.5, label='C column 2')

# Graph settings
plt.xlim(-50, 50)
plt.ylim(-50, 50)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("Vectors from Matrices A (blue), B (green), and C=AB (red)")
plt.show()
```

```python
def visualize_matrix_multiplication(A, B):
    """
    Visualize column vectors of matrices A, B, and their product C = AB (scaled by 1/4).
    
    Parameters:
        A (np.ndarray): A 2x2 matrix
        B (np.ndarray): A 2x2 matrix
    """
    # Validate input shape
    if A.shape != (2, 2) or B.shape != (2, 2):
        raise ValueError("Both A and B must be 2x2 matrices.")
    
    # Matrix multiplication
    C = np.dot(A, B)

    # Create the figure
    plt.figure(figsize=(8, 8))
    plt.axhline(0, color='gray', lw=1)
    plt.axvline(0, color='gray', lw=1)

    # Plot vectors from A
    plt.quiver(0, 0, A[0,0], A[1,0], angles='xy', scale_units='xy', scale=1, color='blue', label='A column 1')
    plt.quiver(0, 0, A[0,1], A[1,1], angles='xy', scale_units='xy', scale=1, color='blue', alpha=0.5, label='A column 2')

    # Plot vectors from B
    plt.quiver(0, 0, B[0,0], B[1,0], angles='xy', scale_units='xy', scale=1, color='green', label='B column 1')
    plt.quiver(0, 0, B[0,1], B[1,1], angles='xy', scale_units='xy', scale=1, color='green', alpha=0.5, label='B column 2')

    # Plot vectors from C
    plt.quiver(0, 0, C[0,0], C[1,0], angles='xy', scale_units='xy', scale=1, color='red', label='C column 1')
    plt.quiver(0, 0, C[0,1], C[1,1], angles='xy', scale_units='xy', scale=1, color='red', alpha=0.5, label='C column 2')

    # Plot settings
    plt.xlim(-50, 50)
    plt.ylim(-50, 50)
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.legend()
    plt.title("Vectors from A (blue), B (green), and C = (red)")
    plt.show()
    print(C)
```
